```python
import pandas as pd
```

```python
import yfinance as yf
```

```python
import plotly.express as px
```

```python
import datetime
```

```python
start = datetime.date(2022,1,1)
```

```python
end = datetime.date.today()
```

```python
ticker = 'AAPL'
```

```python
aapl = yf.download(ticker, start, end)
```

```python
aapl
```

```python
value = 'Close'
```

```python
fig = px.line(aapl, x=aapl.index, y=value, title=f'{ticker} {value}')
```

```python
fig.show()
```

```python
fig.add_scatter(x=aapl.index, y=aapl['Volume'])
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
fig = make_subplots(specs=[[{'secondary_y': True}]])
```

```python
fig.add_trace(
    go.Scatter(x=aapl.index, y=aapl.Close, name='Close'),
    secondary_y=False
)
```

```python
fig.add_trace(
    go.Scatter(x=aapl.index, y=aapl.Volume, name='Volume'),
    secondary_y=True
)
```

```python
fig.update
```

```python
fig = go.Figure()
```

```python
fig.add_scatter(x=aapl.index, y=aapl.Close, name='Close')
```

```python
fig.update_layout(
    title_text=f"AAPL Close",
    title_x=0.5,
    xaxis_title="Date",
    yaxis_title=f"Close",
    legend_title="AAPL",
)
```
